```python
with fitz.open(DIGITIZED_FILE) as doc:
  block_dict = {}

  page_num = 1
  line_num_test = 1

  for page in doc: # Iterate all pages in the document

        file_dict = page.get_text('dict') # Get the page dictionary 

        block = file_dict['blocks'] # Get the block information

        for a in block:   
          if a["type"] == 0:
            for line in a['lines']:
              for span in line['spans']:
                #print(c['text'])
                span["page_num"] = page_num
                span["line_num"] =  line_num_test

              line_num_test += 1

          else :
            a["page_num"] = page_num

        block_dict[page_num] = block # Store in block dictionary

        page_num += 1 # Increase the page value by 1
```

Dataframe creation with all variable extracted for each text line.



*   Font Size

```python
rows = []

for page_num, blocks in block_dict.items():
    for block in blocks:
        if block['type'] == 0:
            for line in block['lines']:
                for span in line['spans']:
                    xmin, ymin, xmax, ymax = list(span['bbox'])
                    font_size = span['size']
                    color = span['color']
                    text = unidecode(span['text'])
                    span_font = span['font']
                    num_page = span["page_num"]
                    line_num = span['line_num']
                    block_num = block['number']
                    is_upper = False
                    is_bold = False 
                    if "bold" in span_font.lower():
                        is_bold = True 
                    if re.sub("[\(\[].*?[\)\]]", "", text).isupper():
                        is_upper = True
                    if text.replace(" ","") !=  "":
                        rows.append((xmin, ymin, xmax, ymax, line_num, block_num, num_page, text, is_upper, is_bold, span_font, font_size, color))
                        

span_df = pd.DataFrame(rows, columns=['xmin','ymin','xmax','ymax', 'line_num', 'block_num', "page_num", 'text', 'is_upper','is_bold','span_font', 'font_size', 'color'])
```

```python
import datetime
#from datetime import datetime


def date(df, test, index):

  """
  Desintegrates list of tuples date into list of date in format mm-yy
  """

  for num, date in enumerate(test): 
    month = date[0].lower()
    year = date[-1]
    if month in month_map:
        month = month_map[month]
        new_date = month + '-' + year
    elif (month.isdigit() and int(month) <= 12) :
        new_date = month + '-' + year
    else :
      new_date = datetime.date.today().strftime("%m-%Y")
    test[num] = new_date

  ##
  #print(test)
  ##

  if len(test) > 1:
    df.loc[index, 'start'] = [test[0]]
    df.loc[index, 'end'] = [test[-1]]
    df.loc[index, 'start'] = [pd.to_datetime(df.loc[index, 'start'],format='%m-%Y').to_period('M')]
    df.loc[index, 'end'] = [pd.to_datetime(df.loc[index, 'end'],format='%m-%Y').to_period('M')]
    df.loc[index, 'Exp_time'] = [(df.loc[index, 'end'].year - df.loc[index, 'start'].year)*12 + (df.loc[index, 'end'].month - df.loc[index, 'start'].month)] 

  else : # SOLAMENTE SI TEXTO TIENE ACTUALIDAD-PRESENTE-PRESENT-ONGOING
    df.loc[index, 'start'] = test[0]
    df.loc[index, 'end'] = np.nan
    df.loc[index, 'start'] = [pd.to_datetime(df.loc[index, 'start'],format='%m-%Y').to_period('M')]
    #df.loc[index, 'end'] = [pd.to_datetime(df.loc[index, 'end'],format='%m-%Y').to_period('M')]
    #df.loc[index, 'Exp_time'] = [(df.loc[index, 'end'].year - df.loc[index, 'start'].year)*12 + (df.loc[index, 'end'].month - df.loc[index, 'start'].month)]

  ##

  

  return df
```

```python
month_map = {
    'january': '1', 'february': '2', 'march': '3', 'april': '4', 'may': '5', 'june': '6',
    'july': '7', 'august': '8', 'september': '9', 'october': '10', 'november': '11', 'december': '12',
    'enero': '1', 'febrero': '2', 'marzo': '3', 'abril': '4', 'mayo': '5', 'junio': '6',
    'julio': '7', 'agosto': '8', 'septiembre': '9', 'octubre': '10', 'noviembre': '11', 'diciembre': '12',
    'jan': '1', 'feb': '2', 'mar': '3', 'apr': '4', 'may': '5', 'jun': '6',
    'jul': '7', 'aug': '8', 'sep': '9', 'oct': '10', 'nov': '11', 'dec': '12',
    'ene': '1', 'feb': '2', 'mar': '3', 'abr': '4', 'may': '5', 'jun': '6',
    'jul': '7', 'ago': '8', 'sep': '9', 'oct': '10', 'nov': '11', 'dic': '12',
}

ongoing = ['present', 'actualidad', 'today', 'presente', 'ongoing', 'currently', 'current', 'hasta la fecha']

symbols = ['-', ',', '(', ')', ':', '.']

min_year = 1900
max_year = 2100
years_av = list(map(lambda x: str(x), list(range(min_year,max_year))))
```

```python
def joining(palabras, keywords):
  
  for index, item in enumerate(palabras):
      concatenated_string = item.lower()
      result = [key for key in keywords if concatenated_string in key]
      contador = 1

      """if index == len(palabras) - 1 or len(palabras) == 1: 
        contador = 0
      else : 
        contador = 1"""

      while result and len(palabras) > 1 and index < len(palabras) - 1 and (index + contador) <= len(palabras) - 1:
        concatenated_string += palabras[index+contador].lower()
        result = [key for key in keywords if concatenated_string in key]
        contador += 1

        if concatenated_string in keywords:
            if index+contador == len(palabras)-1:
              palabras.append(palabras[-1])
            del palabras[index:index+contador]
            palabras.insert(index, concatenated_string)
            
            break

      else:
        concatenated_string = ''

  return palabras
```

```python
def sep_char(text):
  words = text.split()
  
  for i, word in enumerate(words):
        if any(symbol in word for symbol in symbols):
            split_date = re.split('|'.join(re.escape(s) for s in symbols), word)
            words[i:i+1] = split_date

  words = [word for word in words if word != '']

  return words
```

```python
from dateutil import parser
import datetime
import re


def extract_dates(text, min_year, max_year, symbols, ongoing, years_av):
    dates = []
    year = 0
    year_list = []
    month = 0
    month_list = []
    
    words = sep_char(text)
    words = joining(words, list(month_map.keys())[0:25])
    words = joining(words, ongoing)
    words = joining(words, years_av)

    print(words)
    for index, word in enumerate(words):
        if word.isdigit() and (int(word) > min_year and int(word) < max_year):
            year = int(word)
            year_list.append(year)

        elif word.lower() in month_map:
            word = word.lower()
            month = int(month_map[word])
            month_list.append(month)

        elif index < len(words) - 1 and word.isdigit() and int(word) <= 12:
            if words[index+1].isdigit() and (int(words[index+1]) > min_year and int(words[index+1]) < max_year):
                month = int(word)
                month_list.append(month)
                year = int(words[index+1])
                year_list.append(year)


        elif re.findall(r'\b([0-9]{1,2})[-/]([0-9]{4})\b', word, re.IGNORECASE):
            date_parts = word.split("/")
            
            if len(date_parts) > 2: # agregar check para que el numero sea menor o igual a 12
              month = int(date_parts[1])
              year = int(date_parts[-1])
            else : 
              month = int(date_parts[0])
              year = int(date_parts[-1])

            date = parser.parse(word, default=datetime.datetime(year, month, 1, 0, 0))
            dates.append(date)
            year = 0
            month = 0
            
        if year > 0 and month > 0 :
            try:
                date = parser.parse(word, default=datetime.datetime(year, month, 1, 0, 0))
                dates.append(date)
                year = 0
                month = 0
            except ValueError:
                pass
        elif any(word.lower() in ongoing for ongoing in ongoing) and len(dates) > 0 and (len(month_list) == 1  and len(year_list) == 1): #valorar nuevamente las condiciones
                print(word)
                date = datetime.date.today()
                dates.append(date)
        elif len(year_list) <= 2 and len(month_list) == 0 and index == len(words) - 1:
                for item in year_list:
                  date = parser.parse(str(item), default=datetime.datetime(item, 1, 1, 0, 0))
                  dates.append(date)
                
                if len(dates) == 1 and any(word.lower() in ongoing for ongoing in ongoing):
                  dates.append(datetime.date.today())
                print(dates)
        
    return dates
```

```python
def delta_dates(something): 
    years = []
    months = []
    
    #print(something)
    if something :
      for item in something:
        years.append(item.year)
        months.append(item.month)
      delta = (years[-1] - years[0])*12 + (months[-1] - months[0])
    
    else :
      delta = 0

    return delta
```

```python
def assignation(text):

    dates = []
    delta = []
    #len_exp = []
    #indexes = []
    for index, text_line in enumerate(text) : 
      #if extract_dates(text_line):
        #print(text_line)
        dates.append(list(extract_dates(text_line, min_year, max_year, symbols, ongoing, years_av)))

    for item in dates :
      delta.append(delta_dates(item))
      #len_exp.append(len(dates))

    return dates, delta
```

Headers characteristics.

    *   EXPERIENCE

*   These are all in separate lines, therefore if tokenized len of these text would be `len = 2`.
*   Generally speaking most of these header **will have at least one** of the following characteristics:

    *   Upper letters
    *   Higher Font Size
    *   Different font than the rest of the text
    *   Different color (0, means black)

Main Issue with this solution: within every section there is going to be unnecessary text and sometimes it will catch range of dates that are not experiences or education

**Proposal to find dates:** create a function that find lines that comply with at least one of these attributes (1 or more). 

*   Upper letters
*   Higher Font Size
*   Different font than the rest of the text
*   Different color (0, means black)
