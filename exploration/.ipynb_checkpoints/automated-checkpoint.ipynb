```python
with fitz.open(os.path.join(cv_folder, fname)) as doc:
    block_dict = {}
    page_num = 1
    line_num_test = 1
    for page in doc: # Iterate all pages in the document
        file_dict = page.get_text('dict') # Get the page dictionary 
        block = file_dict['blocks'] # Get the block information
        for a in block:
            if a["type"] == 0:
                for line in a['lines']:
                    for span in line['spans']:
                        #print(c['text'])
                        span["page_num"] = page_num
                        span["line_num"] =  line_num_test

                    line_num_test += 1

            else :
                a["page_num"] = page_num

            block_dict[page_num] = block # Store in block dictionary

        page_num += 1 # Increase the page value by 1
```

```python
labels  = pd.DataFrame(columns = ['label','span_start','span_end','text'])
for e in doc2.ents:
    row = {}
    row['label'] = [e.label_]
    row['span_start'] = [e.start_char]
    row['span_end'] = [e.end_char]
    row['text'] = [e.text]
    labels = pd.concat([labels,pd.DataFrame(row)] , ignore_index=False)
labels
```

```python
dates_df.columns = ['_'.join([str(x),str(y)]) for x,y in dates_df.columns]
```

```python
dates_df.columns
```

```python
df_date_to_concat = dates_df[['span_start_0','span_end_1','date_formated_0','date_formated_1','duration_']].rename(columns = {'span_start_0':'span_start', 'span_start_1':'span_end'})
df_date_to_concat['label']='DATE'
df_concat = pd.concat([df_date_to_concat,labels], axis=0, ignore_index=False).sort_values(by = 'span_start')
df_concat['date_formated_0'] = df_concat['date_formated_0'].ffill()
df_concat['date_formated_1'] = df_concat['date_formated_1'].ffill()
df_concat['duration_'] = df_concat['duration_'].ffill()
df_concat.drop_duplicates(['date_formated_0','text'], inplace=True)
df_concat = df_concat[df_concat['label']=='SKILL'].groupby('text').agg({'duration_':'sum','date_formated_0':'min','date_formated_1':'max' }).reset_index()
df_concat['years'] = (df_concat['duration_']/12).round(1)
df_concat['years_threshold'] = pd.cut(df_concat['years'], bins = [0,0.5,1,2,3,4,100], labels = [0,0.5,1,2,3,4], right=False)
df_concat
```

```python
df_scores = pd.read_excel('../data/config.xlsx', sheet_name = 'scores')
df_scores
```

```python
df_concat = df_concat.merge(df_scores, how = 'left', on = 'years_threshold' , validate = 'many_to_one')
df_concat
```
